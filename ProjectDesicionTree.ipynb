```python
import pandas as pd
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = text.split()
    return tokens

# Apply the preprocess_text function to the 'full_text' column
komentar['processed_text']= komentar['full_text'].apply(preprocess_text)

# Access the columns using a list of column names
# Now 'full_text' and 'processed_text' should exist
print(komentar[['full_text', 'processed_text']])
```
